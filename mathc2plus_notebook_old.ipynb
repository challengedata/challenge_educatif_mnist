* Pour récupérer tous les pixels de `x` appartenant aux colonnes `c` à `d-1`, on utilise `x[:,c:d]`; et `x[a:b,:]` pour tous les pixels appartenant aux lignes `a` à `b-1`.

## I.2 Programmation d'un algorithme de classification

### Caractéristique discriminante d’une image 

On commence par définir une **caractéristique discriminante** de référence (ou *benchmark*) d'une image `x`, qui calcule la valeur moyenne de ses pixels. <b><font color="red"> Vous modifierez par la suite cette fonction afin de définir votre propre caractéristique.</font></b>

```python
def caracteristique(x):
    # Initialisation de la somme à zéro
    resultat = 0 
    for i in range(28):  # l'instruction for i in range(N) fait parcourir {0,1,..,N-1} à i
        for j in range(28): 
            resultat = resultat + x[i,j]   # rajout de la valeur du pixel (i,j) à la somme
    return resultat / (28*28)
```

On calcule ensuite cette caractéristique sur toutes les images de `x_train_2`, et on stocke les résultats dans une liste `c_train`. 

Exécutez la cellule suivante qui calcule la liste `c_train`.

```python
# On commence par définir et initialiser la liste c_train (vide au début)
c_train = []

# On la remplit à l'aide d'une boucle for :
for x in x_train_2:
    c_train.append(caracteristique(x))
```

On affiche ci-dessous les dix premières images d'entraînement, leurs caractéristiques calculées par la fonction `caractéristique()`, ainsi que les chiffres associés contenus dans `y_train_2`. 

Exécutez pour cela la cellule suivante.

```python
affichage_dix(x_train_2[:10])
print("c_train =", [float(f"{c:.2f}") for c in c_train[:10]])
print("y_train =", y_train_2[:10].tolist())
```

On affiche ci-dessous les dix images d'entraînement suivantes, leurs caractéristiques, ainsi que les chiffres associés. 

Exécutez pour cela la cellule suivante.

```python
affichage_dix(x_train_2[10:20])
print("c_train =", [float(f"{c:.2f}") for c in c_train[10:20]])
print("y_train =", y_train_2[10:20].tolist())
```

### Algorithme de classification

On définit ensuite notre **algorithme de classification** qui va estimer si une image `x`est un $0$ ou un $1$, suivant que la valeur `c = caractéristique(x)` de sa caractéristique est en-dessous ou au-desssus d'un certain seuil. 

L'algorithme de classification ci-dessous est un benchmark qui s'applique expressément à la caractéristique benchmark de la valeur moyenne des pixels précédemment définie, et utilise un seuil de 46. <b><font color="red"> Vous modifierez par la suite cet algorithme afin de l'adapter à la caractéristique que vous avez choisie. </font></b>

```python
def classification(c):
    # Si la caractéristique de la valeur moyenne des pixels est plus grande que 46, on retourne 0, sinon 1
    if c > 46:
        return 0
    else:
        return 1
```

#### Un algorithme de classification fondé sur les distances aux caractéristiques moyennes

L'algorithme de classification ci-dessous prédit, à partir de la caractéristique `c = caractéristique(x)` d'une image `x`, le chiffre dont la caractéristique moyenne est la plus proche.

Exécutez la cellule suivante pour pouvoir l'utiliser.

```python
def classification(c):
    # On définit d'abord les différentes estimations possibles 
    chiffres = [0,1,4,8] 
    # On calcule le carré des distances entre la caractéristique c et les caractéristiques moyennes
    dist = [(c - moyenne_car)**2 for moyenne_car in c_train_moyennes_par_population]
    # On extrait l'indice minimisant cette distance
    index_min = dist.index(min(dist))
    # On renvoie le chiffre correspondant
    return chiffres[index_min]
```

## II.6 Utiliser deux caractéristiques discriminantes au lieu d'une seule

Les histogrammes précédemment affichés montrent qu'avec quatre classes, il est difficile de trouver une seule caractéristique distinguant bien les quatre chiffres, ce qui induit un taux d'erreur élevé. 

Nous allons voir qu'utiliser un plus grand nombre de caractéristiques discriminantes permet de réduire ce taux d'erreur, avec un algorithme de classification adapté.

Revisualisons quelques images d'entraînement de MNIST-4 afin de réfléchir aux caractéristiques discriminantes pertinentes sur cette base de données.

Exécutez la cellule ci-dessous pour afficher les trentre premières images de la base d'entraînement.

```python
affichage_trente(x_train_4)
```

Nous commençons par définir une paire de caractéristiques discriminantes de référence d'une image `x`, qui calculent la valeur moyenne de ses pixels sur les moitiés supérieure et inférieure de l'image, et sont contenues dans une liste de longueur deux. <b><font color=red> Vous définirez ci-dessous vos deux propres caractéristiques en modifiant la cellule suivante</font></b>.

```python
def deux_caracteristiques(x):
    # Initialisation des deux caractéristiques à 0 : somme[0] représentant la valeur moyenne des pixels sur 
    # la moitié supérieure de l'image, et somme[1] sur la moitié inférieure
    somme = [0, 0] 
    for i in range(14):  # on ne parcourt que les lignes de la moitié supérieure
        for j in range(28): 
            somme[0] = somme[0] + x[i,j]   # rajout de la valeur du pixel (i,j) à la somme
            somme[1] = somme[1] + x[i+14,j]  # décalage de 14 pour parcourir les lignes de la moitié inférieure
    # Normalisation
    somme[0] = somme[0] / (14*28)
    somme[1] = somme[1] / (14*28)
    return somme
```

On calcule ces deux caractéristiques sur toutes les images de `x_train_4`, et on stocke les résultats dans une liste `c_train`. 

Exécutez la cellule suivante qui calcule la liste `c_train`.

```python
# On commence par définir et initialiser la liste c_train (vide au début)
c_train = []

# On la remplit à l'aide d'une boucle for :
for x in x_train_4:
    c_train.append(deux_caracteristiques(x))
```

Nous segmentons ensuite ces deux caractéristiques entre les différentes populations de chiffres de la base de données MNIST-4, et stockons les résultats dans une liste de longueur quatre `c_train_par_population`.

Exécutez la cellule suivante afin de calculer `c_train_par_population`.

```python
chiffres = [0,1,4,8]  
c_train_par_population = [np.array(c_train)[y_train_4==k] for k in chiffres]
```

### Un algorithme de classification fondé sur les distances aux caractéristiques moyennes

#### Distance

Afin de pouvoir utiliser l'algorithme de classification précédemment défini minimisant la distance aux caractéristiques moyennes, il est nécessaire de définir cette distance en deux dimensions. En effet, avec une seule caractéristique `c`, nous calculions la distance aux caractéristiques moyennes avec la ligne de code :

`dist = [(c - moyenne_car)**2 for moyenne_car in c_train_moyennes_par_population]`

qu'il est nécessaire d'adapter à deux caractéristiques.

On peut pour cela utiliser la **distance euclidienne**, dont on rappelle la définition : si $a = (a[0],a[1])$ et $b=(b[0],b[1])$ sont deux points en deux dimensions, alors le carré $d(a,b)^2$ de leur distance euclidienne vaut :

$$d(a,b)^2 = (a[0] - b[0])^2 + (a[1] - b[1])^2$$

Exécutez la cellule suivante afin de définir la distance euclidienne au carré entre deux points `a`et `b`en deux dimensions.

```python
# a et b sont supposés être des points en deux dimensions contenus dans des listes de longueur deux
def distance_carre(a,b):  
    if len(a)!=2 or len(b)!=2:
        raise ValueError("Cette distance euclidienne n'est définie que pour des points en dimension deux")
    return (a[0]-b[0])**2 + (a[1]-b[1])**2
```

On peut à présent redéfinir notre algorithme de classification.

```python
def classification(c):
    # On définit d'abord les différentes estimations possibles 
    chiffres = [0,1,4,8] 
    # On calcule le carré des distances entre la caractéristique c et les caractéristiques moyennes
    dist = [distance_carre(c, moyenne_car) for moyenne_car in c_train_moyennes_par_population]
    # On extrait l'indice minimisant cette distance
    index_min = dist.index(min(dist))
    # On renvoie le chiffre correspondant
    return chiffres[index_min]
```
